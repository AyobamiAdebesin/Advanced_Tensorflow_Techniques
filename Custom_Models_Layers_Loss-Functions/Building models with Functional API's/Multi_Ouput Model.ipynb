```python
import numpy as np
import matplotlib
import time
import os, shutil
import tensorflow as tf
import keras
from tensorflow.keras.models import Model, load_model
from tensorflow.keras.layers import Conv2D, Dense, Flatten, Input
```

```python
def build_model_with_functional():
    input_layer = Input(shape=(28, 28))
    flatten_layer = Flatten()(input_layer)
    first_dense = Dense(128, activation=tf.nn.relu)(flatten_layer)
    output_layer = Dense(10, activation=tf.nn.softmax)(first_dense)
    return Model(inputs=input_layer, outputs=output_layer)

model = build_model_with_functional()
model.summary()
```

```python
#model with multiple outputs
def build_model_multiple_outputs():
    x1 = Input(shape=(None, 8))
    x2 = Dense(128, activation=tf.nn.relu)(x1)
    x3 = Dense(128, activation=tf.nn.relu)(x2)
    
    y1 = Dense(1, activation=tf.nn.softmax)(x3)
    
    x4 = Dense(64, activation=tf.nn.relu)(x3)
    y2 = Dense(1, activation=tf.nn.softmax)(x4)
    
    return Model(inputs=[x1], outputs=[y1, y2])

model2 = build_model_multiple_outputs()
model2.summary()
```
